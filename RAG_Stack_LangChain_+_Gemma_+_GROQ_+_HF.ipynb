```python
import getpass
import os

if "GROQ_API_KEY" not in os.environ:
    os.environ["GROQ_API_KEY"] = getpass.getpass("Provide your GROQ API TOKEN")
```

```python
# Document Loader
from langchain.document_loaders import UnstructuredPDFLoader
loader = UnstructuredPDFLoader(r'/content/drive/MyDrive/AI_ML_Resume_Tools_Trends_2025.pdf')
documents = loader.load()
```

```python
documents
```

```python
import textwrap

def wrap_text_preserve_newlines(text, width=110):
    # Split the input text into lines based on newline characters
    lines = text.split('\n')

    # Wrap each line individually
    wrapped_lines = [textwrap.fill(line, width=width) for line in lines]

    # Join the wrapped lines back together using newline characters
    wrapped_text = '\n'.join(wrapped_lines)

    return wrapped_text
```

```python
print(wrap_text_preserve_newlines(str(documents[0])))
```

```python
# Text Splitter
from langchain.text_splitter import CharacterTextSplitter
text_splitter = CharacterTextSplitter(chunk_size=1000, chunk_overlap=0)
docs = text_splitter.split_documents(documents)
```

```python
docs[1]
```

```python
len(docs)
```

```python
from langchain.embeddings import HuggingFaceEmbeddings

embeddings = HuggingFaceEmbeddings()
```

```python
from langchain.vectorstores import FAISS

db = FAISS.from_documents(docs, embeddings)
```

```python
query = "List out machine learning algorithms"
docs = db.similarity_search(query)
```

```python
print(wrap_text_preserve_newlines(str(docs[0].page_content)))
```

```python
import langchain_groq
from langchain_groq import ChatGroq

GROQ_LLM = ChatGroq(
            api_key=os.getenv("GROQ_API_KEY"),
            model="gemma2-9b-it"
        )
```

```python
from langchain.chains.question_answering import load_qa_chain

chain = load_qa_chain(GROQ_LLM, chain_type="stuff")
```

```python
query = "HOW IS INDIAN ECONOMY"
docs = db.similarity_search(query)
chain.run(input_documents=docs, question=query)
```

```python
query = "List out the technologies related to Generative AI"
docs = db.similarity_search(query)
chain.run(input_documents=docs, question=query)
```

```python
query = "List out the technologies related to Generative AI"

# Retrieve similar documents
docs = db.similarity_search(query)

# Run the chain
response = chain.run(input_documents=docs, question=query)

# Neatly print everything
print("\n" + "="*60)
print("Query:")
print(query)

print("\n Retrieved Documents:")
for i, doc in enumerate(docs, 1):
    print(f"\n--- Document {i} ---")
    print(doc.page_content.strip())

print("\n Response:")
print(response)
print("="*60 + "\n")
```
